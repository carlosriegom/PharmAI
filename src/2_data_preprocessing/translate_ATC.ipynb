#### **2.2 Dividir el `name_L5.json` en varios json más pequeños para simplificar la traducción**

```python
# Dividir en 10 jsons el name_L5.json
def split_json_file(input_file, output_dir, num_files):
    # Cargar el archivo JSON
    with open(input_file, 'r') as f:
        data = json.load(f)

    # Calcular el tamaño de cada fragmento
    chunk_size = len(data) // num_files

    # Crear la carpeta de salida si no existe
    os.makedirs(output_dir, exist_ok=True)

    # Dividir y guardar los fragmentos
    for i in range(num_files):
        start_index = i * chunk_size
        end_index = (i + 1) * chunk_size if i != num_files - 1 else len(data)
        chunk = dict(list(data.items())[start_index:end_index])
        
        output_file = os.path.join(output_dir, f"name_L5_part_{i + 1}.json")
        with open(output_file, 'w') as f:
            json.dump(chunk, f, indent=4)

# Ruta del archivo JSON de entrada
input_file = os.path.join(ROOT_DIR, "niveles_sin_traducir", "name_L5.json")

# Ruta del directorio de salida
output_dir = os.path.join(ROOT_DIR, "niveles_sin_traducir", "name_L5_parts")

# Número de archivos JSON a crear
num_files = 10

# Llamar a la función para dividir el archivo JSON
split_json_file(input_file, output_dir, num_files)

# Comprobar que los archivos json se han creado correctamente
print("Archivos json creados correctamente:")
for i in range(num_files):
    print(f"name_L5_part_{i + 1}.json:", os.path.exists(os.path.join(output_dir, f"name_L5_part_{i + 1}.json")))
```

#### **2.4 Generar la `Tabla_ATC_traducida.xlsx`**

```python
# Cargar Tabla_ATC.xlsx
atc_df = pd.read_excel("../../assets/Tabla_ATC.xlsx")

# Ver las columnas de la tabla
print("Las columnas de la tabla son:")
print(atc_df.columns)

# Eliminar las columnas que no son necesarias
atc_df = atc_df[['ATC code_L1', 'name_L1', 'ATC code_L2', 'name_L2', 'ATC code_L3', 'name_L3', 'ATC code_L4', 'name_L4', 'ATC code_L5', 'Name_L5']]

# Cambiar el nombre de las columna 'Name_L5' a 'name_L5'
atc_df.rename(columns={'Name_L5': 'name_L5'}, inplace=True)

# Ver las columnas de la tabla
print("\nLas columnas de la tabla son:")
print(atc_df.columns)
```

```python
# Comvertir los campos de los nombres a minúsculas
atc_df['name_L1'] = atc_df['name_L1'].str.lower()
atc_df['name_L2'] = atc_df['name_L2'].str.lower()
atc_df['name_L3'] = atc_df['name_L3'].str.lower()
atc_df['name_L4'] = atc_df['name_L4'].str.lower()
atc_df['name_L5'] = atc_df['name_L5'].str.lower()

# Hacer un join entre cada columna "name_LX" del dataframe y el diccionario traducido correspondiente, para sustituir el nombre en inglés por el nombre en español
def translate_column(df, translated_dicts_directory):
    for column in ["name_L1", "name_L2", "name_L3", "name_L4", "name_L5"]:
        # Cargar el diccionario traducido
        translated_dict = load_json_dict(os.path.join(translated_dicts_directory, f"{column}_traducido.json"))
        
        # Hacer un join entre la columna y el diccionario traducido
        df[column] = df[column].map(translated_dict).fillna(df[column])
    
    return df

# Traducir cada columna
translated_dicts_directory = os.path.join("..", "..", "data", "outputs", "2_data_preprocessing", "niveles_traducidos")
atc_df_traducido = translate_column(atc_df, translated_dicts_directory)

# Comprobar que la longitud de la tabla traducida es la misma que la original
print("Longitud de la tabla traducida:")
print("atc_df:", len(atc_df))
print("atc_df_traducido:", len(atc_df_traducido))

# Ver los primeros registros de la tabla traducida
print("Los primeros registros de la tabla traducida son:")
display(atc_df_traducido.head())
```

```python
# Guardar la tabla traducida en un archivo Excel
output_file = os.path.join("..", "..", "assets", "Tabla_ATC_traducida.xlsx")
atc_df_traducido.to_excel(output_file, index=False)
print("Archivo Excel creado correctamente:")
print("Tabla_ATC_traducida.xlsx:", os.path.exists(output_file))
```
